```python
import os

from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(data='/content/gdrive/MyDrive/Thai-dessert/data.yaml', epochs=50, imgsz=640)
```

```python
# Load the trained model
model = YOLO('/content/gdrive/MyDrive/Thai-dessert/model/runs/detect/train2/weights/best.pt')

# Define path to the image file
source = '/content/gdrive/MyDrive/Thai-dessert/Pic4Test/ทองหยิบ.jpg'

# Run inference on the source
results = model.predict(source)  # list of Results objects

# Get the Result object for the first (and only) image
result = results[0]

# Display the image with bounding boxes
result.save()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Save the image with bounding boxes
result.save()

# The image will be saved in a 'runs/detect' directory
saved_image_path = '/content/results_ทองหยิบ.jpg'  # adjust the path as needed

# Display the image
img = mpimg.imread(saved_image_path)
imgplot = plt.imshow(img)
plt.show()
```
